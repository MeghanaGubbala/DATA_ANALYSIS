```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("/content/Temperature.txt",sep=" ")
print(df)
print("\n The average temperature in 50 days is:")
avg=df['Temperature'].mean()
print(avg)
print("\n The lowest temperature recorded is:")
print(df['Temperature'].min())
print("\n The highest temperature recorded is:")
print(df['Temperature'].max())
day=0
for i in df['Temperature']:
  if (i>avg):
    day=day+1
print("\n Number of days where the temperature exceeded treshold is:")
plt.plot(df['Days'],df['Temperature'],color='black')
plt.title("Tempetature for 50 days")
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.show()
```
